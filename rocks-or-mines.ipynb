```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
sonar_data = pd.read_csv('../input/rock-or-mine-classification/ROCK_OR_MINE.csv', header = None)
```

```python
sonar_data.head()
```

```python
#number of rows and coloum
sonar_data.shape
```

```python
# gives statistical data
sonar_data.describe()
```

```python
#gives the number of rocks and mine datas
#we write 60 because R and M Lies on the 60 the column
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
#seperating data and labels
s = sonar_data.drop(columns = 60  , axis = 1)
n = sonar_data[60]
```

```python
print(s)
print(n)
```

```python
#spliting train and test data
s_train , s_test , n_train , n_test = train_test_split(s,n , test_size = 0.1 , stratify = n , random_state = 42 )
```

```python
print(s_train.shape , s_test.shape)
```

```python
#model training
model = LogisticRegression()
```

```python
#training logistic regression
model.fit(s_train , n_train)
```

```python
#accuracy check on training data
s_train_prediction = model.predict(s_train)
training_data_accuracy = accuracy_score(s_train_prediction , n_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy check on testing data
n_test_prediction = model.predict(s_test)
testing_data_accuracy = accuracy_score(n_test_prediction, n_test)

```

```python
print(testing_data_accuracy)
```
